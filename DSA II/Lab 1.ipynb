### printA function --> will prinit from 1 upto n

```python
def printA(n):
    if n <= 0:
        print('n must be greater or equal to 1')
    elif n == 1:
        print(1)
    else:
        printA(n - 1)  # Recursive call
        print(n)       # Print after the recursive call

printA(5)
```

### Calculate the sum of numbers from 1 to n

```python
def sum_n(n):
    if n <= 0:
        print('n must be greater or equal to 1')
    elif n == 1:
        return 1
    else:
        return sum_n(n-1) + n
        
sum_n(5)   
```

### Calculate the factorial of n

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)

factorial(5)
```

### Calculate the sum of digits of a given number n

```python
def sum_digit_n(n):
    if n < 10:
        return n
    else:
        return n % 10 + sum_digit_n(n//10)

sum_digit_n(123)
```

### Count the number fo digits of a given number of n

```python
def count_n(n):
    n = abs(n)
    if n == 0:
        return 0
    else:
        return 1 + count_n(n // 10)

count_n(-1290)
        
```

### Calculate the nth term of Fibonacci series

```python
def nth_fibo(n):
    if n < 0:
        print('n must be greater or equal to 1')
    elif n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return nth_fibo(n-1) + nth_fibo(n-2)

nth_fibo(10)
```

### Calculate a to the power b

```python
def power(a, b):
    if b == 0:
        return 1
    elif b < 0:
        return 1 / power(a, -b)
    else:
        return a * power(a, b - 1)

power(9, 3)
```

```python
def mypow(a, b):
    if a == 0:
        return "a can't be zero"
    if b == 0:
        return 1
    v = mypow(a, b//2)
    if b % 2 == 0:
        return v * v
    else:
        return v * v * a

mypow(3, 3)
```

### Print the array elements

```python
def print_array(arr, index=0):
    if index >= len(arr):
        return
    print(arr[index])
    print_array(arr, index + 1)

print_array([2, 3, 4, 5, 6])
```

### Find the largest elemnet of a given array

```python
def find_max(arr, index=0):
    if index == len(arr) - 1:
        return arr[index]  
    max_rest = find_max(arr, index + 1)  
    if arr[index] > max_rest:
        return arr[index]
    else:
        return max_rest

find_max([1, 2, 100])
```

### Find the largest and smallest element of a given array

```python
def find_max_min(arr, index=0):
    if index == len(arr) - 1:
        return arr[index], arr[index]

    current_max, current_min = find_max_min(arr, index + 1)

    if arr[index] > current_max:
        current_max = arr[index]
    if arr[index] < current_min:
        current_min = arr[index]

    return current_max, current_min


find_max_min([1, 34, -1])
```
