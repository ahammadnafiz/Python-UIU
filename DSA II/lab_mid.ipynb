# Problem 1: Factorial Calculator

```python
def problem1():
    t = int(input())
    for _ in range(t):
        n = int(input())
        factorial = 1
        for i in range(1, n + 1):
            factorial *= i
        print(factorial)
```

# Problem 2: Matrix Filling with Recursive Rules

```python
def problem2():
    """
    - a[i][j] = m if i=0 and j=0
    - a[i][j] = a[i-1][j] + k if i>j
    - a[i][j] = a[i][j-1] - 1 if i>j
    - a[i][j] = a[i][j-1] - 1 if i<j
    """
    n, m, k = map(int, input().split())
    
    # Create matrix
    matrix = [[0] * n for _ in range(n)]
    
    # Fill the matrix according to rules
    for i in range(n):
        for j in range(n):
            if i == 0 and j == 0:
                matrix[i][j] = m
            elif i > j:
                matrix[i][j] = matrix[i-1][j] + k
            elif i == j:
                if i == 0:
                    matrix[i][j] = m
                else:
                    matrix[i][j] = matrix[i][j-1] - 1
            else:  # i < j
                matrix[i][j] = matrix[i][j-1] - 1
    
    # Print the matrix
    for row in matrix:
        print(' '.join(map(str, row)))
```

# Problem 3: Matrix Self-Multiplication

```python
def problem3():
    n = int(input())
    
    # Original matrix A
    result = [[1, 2], [3, 4]]
    
    # Multiply the matrix by itself (N-1) times
    for _ in range(n - 1):
        # Matrix multiplication: result = result * A
        new_result = [[0, 0], [0, 0]]
        a = [[1, 2], [3, 4]]
        
        for i in range(2):
            for j in range(2):
                for k in range(2):
                    new_result[i][j] += result[i][k] * a[k][j]
        
        result = new_result
    
    # Print the first element
    print(result[0][0])
```

# Problem 4: Integer Square Root using Binary Search

```python
def problem4():
    t = int(input())
    for _ in range(t):
        n = int(input())
        
        # Binary search for square root
        if n == 0:
            print(0)
            continue
        
        left = 1
        right = n
        result = 1
        
        while left <= right:
            mid = (left + right) // 2
            
            # Check if mid * mid equals n
            if mid * mid == n:
                result = mid
                break
            elif mid * mid < n:
                # mid could be the answer, but check for larger values
                result = mid
                left = mid + 1
            else:
                # mid * mid > n, so search in left half
                right = mid - 1
        
        print(result)
```

# Problem 5: Beautifully Closed Markers

```python
def problem5():
    # Read n, m (number of markers and caps)
    n, m = map(int, input().split())
    
    # Read markers
    markers = []
    for _ in range(n):
        x, y = map(int, input().split())
        markers.append((x, y))
    
    # Read caps
    caps = []
    for _ in range(m):
        a, b = map(int, input().split())
        caps.append((a, b))
    
    # Count available caps for each marker specification
    from collections import Counter
    cap_count = Counter(caps)
    
    # Try to close markers optimally
    closed_beautifully = 0
    
    # For each marker, check if there's a matching cap
    used_caps = []
    for marker in markers:
        if marker in cap_count and cap_count[marker] > 0:
            closed_beautifully += 1
            cap_count[marker] -= 1
            used_caps.append(marker)
    
    # Total closed markers = beautifully closed + remaining markers with remaining caps
    remaining_markers = n - closed_beautifully
    remaining_caps = sum(cap_count.values())
    total_closed = closed_beautifully + min(remaining_markers, remaining_caps)
    
    print(total_closed, closed_beautifully)
```
