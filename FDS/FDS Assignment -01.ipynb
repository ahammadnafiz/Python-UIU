```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'EmployeeID': [1, 2, 3, 4, 5, 6, 7, 8],
    'Department': ['HR', 'Engineering', 'Engineering', 'HR', 'Marketing', 'Marketing', 'HR', 'Engineering'],
    'Age': [35, 30, 32, 40, 28, 27, 45, 29],
    'Salary': [50000, 60000, 65000, 55000, 48000, 47000, 53000, 61000]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
grouped = df.groupby('Department')
```

```python
grouped['Age'].mean()
```

```python
grouped['Salary'].mean()
```

```python
def top_two_employee(department):
    return department.sort_values('Salary', ascending=False)['Salary'].head(2)
```

```python
top_two = df.groupby('Department').apply(top_two_employee, include_groups=False)
```

```python
top_two
```

```python
def top_two_employee(department):
    return department.nlargest(2, 'Salary')
top_two = df.groupby('Department').apply(top_two_employee, include_groups=False)
```

```python
top_two
```

```python
data= {
 'Age': [25, 30, 35, 40, 45],
 'Income': [50000, 60000, 75000, 90000, 80000],
 'Education': [12, 16, 14, 18, 15],
 'Savings': [20000, 25000, 30000, 40000, 35000],
 'Spending': [15000, 18000, 20000, 22000, 21000],
 'Exercise Hours': [3, 2, 4, 1, 3]
 }
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data=df.corr(), annot=True, linewidths=0.8, fmt='.2f')

plt.title('Correlation heatmap  of the Dataset', fontsize = 16)
plt.show()
```

### 1. **Strong Positive Correlations**:
   - **Age vs Spending (0.9117)**: A very high positive correlation, suggesting that as age increases, spending also increases significantly.
   - **Age vs Savings (0.9000)**: Another strong positive correlation, indicating that older individuals tend to have more savings.
   - **Income vs Savings (0.9901)**: This near-perfect positive correlation suggests that higher income is strongly associated with higher savings.
   - **Income vs Spending (0.9817)**: This suggests that individuals with higher incomes tend to spend more.
   - **Income vs Age (0.8911)**: Strong positive correlation, showing that income tends to rise as people get older.
   - **Savings vs Spending (0.9687)**: A very high correlation, implying that people with more savings tend to spend more.

### 2. **Moderate Positive Correlations**:
   - **Education vs Income (0.7351)**: This suggests a moderate positive relationship, indicating that higher education levels are somewhat associated with higher income.
   - **Education vs Savings (0.7778)**: Theres a moderate correlation between education and savings, implying that people with more education tend to save more.
   - **Education vs Spending (0.7655)**: A moderate correlation indicates that higher education is somewhat linked to higher spending.

### 3. **Strong Negative Correlations**:
   - **Education vs Exercise Hours (-0.7845)**: A strong negative correlation, indicating that individuals with higher education tend to exercise less.

### 4. **Moderate Negative Correlations**:
   - **Exercise Hours vs Savings (-0.4160)**: This suggests a moderate negative correlation, implying that individuals who exercise more tend to save less.
   - **Exercise Hours vs Income (-0.3158)**: This negative correlation suggests that individuals with higher income might be exercising less, but the relationship is weaker.
   - **Exercise Hours vs Spending (-0.2845)**: This weak negative correlation indicates that higher spending may be slightly associated with fewer hours of exercise.
   - **Exercise Hours vs Age (-0.1387)**: The negative correlation between age and exercise hours is weak, meaning older individuals tend to exercise a little less, but the relationship is not strong.

### 5. **Weak Correlations**:
   - **Age vs Education (0.5657)**: This shows a moderate to weak correlation, suggesting that as people age, there is a slight tendency for higher education levels, but the relationship is not very strong.

### Summary:
- **Strong Positive Relationships** are seen among variables like age, income, savings, and spending. As individuals get older and earn more, they tend to save and spend more.
- **Strong Negative Correlation** between education and exercise hours suggests that individuals with higher education tend to spend less time exercising.
- **Moderate to Weak Relationships** exist between education, age, and exercise hours, implying that the relationships between these factors are present but not particularly strong.
