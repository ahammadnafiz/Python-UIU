```python
import requests
from bs4 import BeautifulSoup

response = requests.get('https://news.ycombinator.com/')
soup = BeautifulSoup(response.text, 'html.parser')
stories = soup.find_all('span', class_='titleline')

print("Top Hacker News stories:")
for i, story in enumerate(stories[:5], 1):
    title = story.find('a').text
    link = story.find('a')['href']
    print(f"{i}. {title}")
    print(f"   Link: {link}")
```

### Sending JSON data to JSONPlaceholder

```python
import json

# Data to send as JSON
post_data = {
    'title': 'My New Post',
    'body': 'This is the content of my post',
    'userId': 1
}

# Send POST request with JSON data
headers = {'Content-Type': 'application/json'}
response = requests.post(
    'https://jsonplaceholder.typicode.com/posts',
    data=json.dumps(post_data),
    headers=headers
)

# Alternative way to send JSON data
# response = requests.post(
#     'https://jsonplaceholder.typicode.com/posts',
#     json=post_data  # requests automatically handles JSON serialization
# )

print(f"Status code: {response.status_code}")
print(f"Response: {response.json()}")
```

### Form submission

```python
# Contact form submission example
form_data = {
    'name': 'John Doe',
    'email': 'john@example.com',
    'message': 'Hello, this is a test message!'
}

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
}

response = requests.post('https://httpbin.org/post', data=form_data, headers=headers)
print(f"Status code: {response.status_code}")
print(response.json())
```

```python
response.json()['headers']['Content-Type'].split('/')[-1]
```

### File upload to a server

```python
# File upload example
files = {
    'file': ('report.csv', open('report.csv', 'rb'), 'text/csv')
}

response = requests.post('https://httpbin.org/post', files=files)
print(f"Status code: {response.status_code}")
print(response.json())
```
