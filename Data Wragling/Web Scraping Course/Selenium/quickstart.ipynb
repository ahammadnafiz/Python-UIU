```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time

import undetected_chromedriver as uc
```

```python
# Launch Chrome in undetected mode
chrome_driver = uc.Chrome()

# Chrome Service
# chrome_service = ChromeService()
# chrome_driver = webdriver.Chrome(service=chrome_service)

# Open Google
chrome_driver.get("https://www.google.com")

# Find search box and enter a query
search_box = chrome_driver.find_element(By.NAME, "q")  # Fixed variable name here
search_box.send_keys("Selenium with undetected_chromedriver")
time.sleep(2)
search_box.send_keys(Keys.RETURN)

print(chrome_driver.title)
print(chrome_driver.current_url)
chrome_driver.save_screenshot("screenshot.png")

# Wait to see results
time.sleep(5)

# Close browser
chrome_driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service as EdgeService
from webdriver_manager.microsoft import EdgeChromiumDriverManager

# Edge Service
edge_service = EdgeService(EdgeChromiumDriverManager().install())
edge_driver = webdriver.Edge(service=edge_service)

# Open Google
edge_driver.get("https://duckduckgo.com/")

# Find search box and enter a query
search_box = edge_driver.find_element(By.NAME, "q")  # Fixed variable name here
search_box.send_keys("Ahammad Nafiz")
time.sleep(2)
search_box.send_keys(Keys.RETURN)

print(edge_driver.title)
print(edge_driver.current_url)
# edge_driver.save_screenshot("screenshot.png")

# Wait to see results
time.sleep(5)

# Close browser
edge_driver.quit()
```
