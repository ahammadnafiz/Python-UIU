### Dropdown & Multiselect

```python
from selenium.webdriver.support.ui import Select
```

```python
driver = uc.Chrome()

url = "https://miniclip.com/careers/vacancies"
driver.get(url)
time.sleep(2)

location_xpath = '//*[@id="__layout"]/div/div/section[2]/div/fieldset[2]/select'
department_xpath = '//*[@id="__layout"]/div/div/section[2]/div/fieldset[3]/select'

location_dropdown = Select(driver.find_element(By.XPATH, location_xpath))
location_dropdown.select_by_visible_text("London, United Kingdom")
time.sleep(2)

department_dropdown = Select(driver.find_element(By.XPATH, department_xpath))
department_dropdown.select_by_visible_text("Product Analytics")

data_analyst_xpath = '//*[@id="__layout"]/div/div/section[3]/div[2]/ul/li/span[1]/a[2]'
data_analyst_click = driver.find_element(By.XPATH, data_analyst_xpath).click()
time.sleep(2)

apply_xpath = '//*[@id="content"]/div/div[2]/div/div[1]/div'
apply_click = driver.find_element(By.XPATH, apply_xpath).click()
time.sleep(2)

time.sleep(20)
driver.quit()
```

```python
driver = uc.Chrome()

url = 'https://demoqa.com/select-menu'
driver.get(url)
time.sleep(2)

cars_element = driver.find_element(By.XPATH, '//*[@id="cars"]')

cars_ms = Select(cars_element)

time.sleep(2)

# selecting options

cars_ms.select_by_index(1)
time.sleep(1)

cars_ms.select_by_visible_text("Opel")
time.sleep(1)

cars_ms.select_by_visible_text("Audi")
time.sleep(1)

# deselecting options

cars_ms.deselect_by_index(2)
time.sleep(1)

cars_ms.deselect_all()
time.sleep(3)

driver.quit()
```
